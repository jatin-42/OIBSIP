```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Unemployment in India.csv')
```

```python
data.isnull().sum()
```

```python
print(data.describe())
```

```python
print(data.head())
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Region', y=' Estimated Unemployment Rate (%)', data=data)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Region')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['number'])
average_unemployment_rate = data[' Estimated Unemployment Rate (%)'].mean()
print(f'Average Unemployment Rate: {average_unemployment_rate:.2f}%')
```

```python
highest_rate_year = data.loc[data[' Estimated Unemployment Rate (%)'].idxmax(), 'Region']
lowest_rate_year = data.loc[data[' Estimated Unemployment Rate (%)'].idxmin(), 'Region']
print(f'Highest Unemployment Rate Year: {highest_rate_year}')
print(f'Lowest Unemployment Rate Year: {lowest_rate_year}')
```

```python
correlation_matrix = numeric_data.corr()
print(correlation_matrix)
```
